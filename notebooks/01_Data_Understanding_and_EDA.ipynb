# 📌 Key Business Insights

## Data Cleaning

* Identified blank values in the `TotalCharges` column.
* Verified that all blank `TotalCharges` values belonged to customers with `tenure = 0`.
* Replaced blank values with `0` and converted `TotalCharges` to a numeric data type.
* Determined that `customerID` is only a unique identifier and should be excluded from model training.

---

## Customer Churn Insights

### 1. Contract Type

* **Month-to-month** customers have the highest churn rate (approximately **42%**).
* Customers with **One-year** and **Two-year** contracts are significantly less likely to churn.
* Contract type appears to be one of the strongest indicators of customer churn.

### 2. Customer Tenure

* Created tenure groups using quartiles to better understand customer behavior.
* Customers in the **first tenure quartile (0–9 months)** have the highest churn rate (approximately **49%**).
* Customers in the **last tenure quartile (56–72 months)** have the lowest churn rate (approximately **7%**).
* This suggests that the initial months are the most critical for customer retention.

### 3. Payment Method

* **Electronic Check** customers exhibit the highest churn rate (approximately **45%**).
* Customers using **Credit Card (Automatic)**, **Bank Transfer (Automatic)**, and **Mailed Check** have considerably lower churn rates.

### 4. Senior Citizens

* Senior citizens represent a smaller portion of the customer base but have a noticeably higher churn rate than non-senior customers.

### 5. Dependents

* Customers **without dependents** have a substantially higher churn rate (approximately **31%**) than customers **with dependents** (approximately **15%**).
* This represents a strong association with churn but does not necessarily imply causation.

### 6. Monthly Charges

* Customers in the higher monthly charge categories experience higher churn rates.
* This indicates a positive association between monthly charges and churn, although additional factors may also contribute.

### 7. Gender

* Male and female customers exhibit very similar churn rates.
* Gender does not appear to have a significant relationship with customer churn.

### 8. Paperless Billing

* Customers using paperless billing show higher churn.
* However, further analysis suggests that part of this relationship may be influenced by other variables such as contract type.

---

# Feature Engineering

## Tenure Groups

* Created `tenure_group` using quartiles to analyze churn across different stages of the customer lifecycle.

## Monthly Charge Groups

* Created `MCharges` to categorize customers based on their monthly charges for easier exploratory analysis.

## Additional Service Count

* Engineered a new feature named `Additional_Service_Count` by counting the number of additional services subscribed to by each customer:

  * Online Security
  * Online Backup
  * Device Protection
  * Tech Support
  * Streaming TV
  * Streaming Movies

### Findings

* Initially, customers with more additional services appeared to have lower churn.
* Further analysis showed that this relationship weakened after controlling for **Contract Type**.
* This suggests that the apparent relationship was largely influenced by contract type rather than the number of additional services alone.

---

# Overall Observations

* Customer churn is associated with several customer characteristics rather than a single factor.
* Contract type, tenure, payment method, and monthly charges appear to be among the strongest indicators of churn.
* Throughout the analysis, care was taken to distinguish **association** from **causation**, recognizing that observed relationships may be influenced by interactions between multiple variables.

---

# Conclusion

The exploratory data analysis provided a comprehensive understanding of customer behavior and identified several factors associated with churn.
